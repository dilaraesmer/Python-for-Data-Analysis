```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize' : (20,10)})
```

```python
data = pd.read_csv('/content/Melbourne_housing.csv')
data.head()
```

```python
object_columns = data.select_dtypes('object').columns

object_columns = np.append(object_columns, 'Car')
object_columns = np.append(object_columns, 'Bathroom')

print(object_columns)
```

```python
for colname in object_columns:
  data[colname] = data[colname].astype('category')

  data.info()
```

```python
duplicate = data.duplicated()
print(duplicate.sum())
```

```python
data[duplicate]
```

```python
data.drop_duplicates(inplace=True, keep='first')
data.duplicated().sum()
```

```python
data.describe().T
```

```python
data.loc[data['Landsize'] <= 0, 'Landsize'] = np.nan
data.loc[data['BuildingArea'] <= 0, 'BuildingArea'] = np.nan
```

```python
data.drop(data.loc[data['YearBuilt'] < 1835].index, inplace=True)
```

```python
data['Age'] = 2021 - data['YearBuilt']
data.drop(data.loc[data["Age"]<0].index, inplace=True)
```

```python
data.head()
```

```python
data = data.drop(['YearBuilt'], axis=1)
```

```python
data.head()
```

```python
numerical_columns = data.select_dtypes("float64","int64").columns
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data.isnull().sum()/len(data)*100
```

```python
categorical_columns = data.select_dtypes("category")
categorical_columns
```

```python
for column in categorical_columns:
  data[column] = data[column].fillna(data[column].mode().iloc[0])
```

```python
data.isnull().sum()
```

```python
num_bins = 50
n, bins, patches = plt.hist(data.Price, num_bins, color='b', alpha=0,5, histtype ='bar', ec='black')
plt.ylabel ('Frequency')
plt.xlabel ('Price ($)')
plt.xlim([0, 6000000])

plt.title('Histogram House Prices')
plt.show()
```
